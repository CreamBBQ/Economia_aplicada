<center> <H1> OFERTA DE TRABAJO </H1> <center>

La fuerza de tabajo o personas activas se compone de todas las personas empleadas o desempleadas. Para ser considerado desempleados, las personas deben: 

- Estar sin trabajo 
- Estar disponible para trabajar 
- Estar buscando trabajo 

La tasa de empleo es la relación entre el número de personas ocupadas en la población. 

La tasa de participación es la relación entre las personas activas y la población de un grupo de edad, por ejemplo, las personas en edad de trabajador (generalmente más de 15 años de edad). La tasa de desempleo es la relación entre el número de personas desempleadas sobre los activos. 

Ocupado: si trabajo al menos una hora la semana anterior a la encuentas (polémico). Se consideran también las personas que habitualmente tienen un empleo pero por condiciones del trabajo no tuvo que ir. Si atiende un kiosko familiar sin recibir ingresos se considera empleado **porque está vinculado a un mercado**, cosa que no pasa, por ejemplo, en las amas de casa. 

Desocupado: es dificil de específicar si están activamente trabajo. Es una definición, por lo menos, ambigua. La encuesta toma si se buscó trabajo en el último mes (esto arbitrario y cambia mucho los resultados). El encuestado debe estar disponible para trabajar. 

Tasa de actividad = Activos / Población 

Tasa de empleo = Ocupados / Población 

Tasa de desempleo = Desocupados / Activos 

Tasa de subocupación = Subocupados / Activos 

Notar que la tasa de desempleo no es el único indicador del dinamismo de la oferta de trabajo. El desocupado puede no estar buscando activamente empleo por transferencias u otras condiciones propias de cada lugar. La relación que hay que establecer es que la actividad también puede estar moviendose entre regiones (por la pandemia por ejemplo). Un mejor indicador podría ser la tasa de empleo formal-privada. 

Entre los ocupados la insercción ocupacional es variada y se puede clasificar como 
- Asalariados (aproximadamente 75% de los ocupados): asalariados formales (50%) y asalariados informales (25%) 
-Independientes (25%): cuentrapropistas (20%) y empleadores (5%)

Resumen de hechos empiritcos: 

En el corto plazo puede haber reacción de la cantidad de trabajo por precio o incentivos 

En el largo plazo: 

- El incremiento de salarios reales en el tiempo fue acompañado de una caída de las horas trabajadas 
- Hubo cambios sustanciales en la tasa de actividad por género 
- Hubo cambios sustanciales en la asignación del tiempo y fuertes diferencias por género 







##### Teoría Neoclásica de la oferta de trabajo 

El problema del trabajador para su oferta de trabajo es un problema del consumidor con las siguientes particularidades 

- Dotaciones ($L_0$, R)
- Restricciones en las horas (no se pueden consumir más de $L_0$ horas)
- Restricciones presupuestarias no necesariamente rectas (por ejemplo, veremos más adelante la función de producción de horas en el hogar)

Repasemos lo que sabemos de la teoría del consumidor con dos bienes $x_1$ y $x_2$ con dotaciones $\omega_1$ y $\omega_2$ con precios $p_1$ y $p_2$. La solución de los problemas de dotaciones: 

$\displaystyle \max_{x_1, x_2} u(x_1, x_2) \space \space \space \text{s.a} \space \space \space p_1 x_1 + p_2 x_2 = p_1 \omega_1 + p_2 \omega_2$

es un vector de demandas marshallianas $x^M(p, \omega)$ y utilidad indirecta $v(p, \omega)$ que verifica: 

- La propiedad TMS es igual a razón de precios en uan solución interior con preferencias convexas: $$\dfrac{\partial \frac{u(x_1^M, x_2^M)}{\partial x_1}}{\partial \frac{u(x_1^M, x_2^M)}{\partial x_2}} = \dfrac{p_1}{p_2}$$

- Es indéntica a la solución de la demandas marshalliana (con un presupuesto fijo) evaluada en la dotación $$ x_i^M(p, \omega) = x_i^M(p,p \omega)

- Satisface la ecuación de Slutsky: $$ \dfrac{\partial x_i^M(p, \omega)}{\partial p_j} = \dfrac{\partial x_i^C}{\partial p_j} (p, v(p, p\omega)) + \dfrac{\partial x_i^C}{\partial m} (\omega_i- x_j^M) $$

- Define la función de oferta neta como $$s_i(p,\omega)= \left\{ \begin{array}{lcc}
              \omega_i - x_i^M  & si  & \omega_i \geq x_i^M\\
             \\ 0  & & \text{en otros casos}\
             \end{array}
   \right.$$

- Existe un precio relativo a partir del cual la oferta neta es cero, y surge de $x_i^M(p^A, \omega) = \omega_i$

El problema también podría ser la maximización de la utildad con dos bienes: consumo y ocio, sujeto a que $C \leq wh + R_0$ Donde wh es el salario por horas trabajadas y $R_0$ sería una renta que tiene la persona en cuestión. De esta forma la condición de primer orden es la tasa marginal de sustitución entre consumo y ocio igualada al salario (normalizamos el precio del consumo a 1). Habrá un salario de reserva $w_A$ a partir del cual el individuo está indiferente entre trabajar o no trabajar (hacer uso de sus dotaciones o sacrificar ocio para aumentar el consumo). El salario ejerce dos efectos distintos sobre la ofera de trabajo influyendo:

##### Producción doméstica 

(imaginemos que los agentes pueden consumir, por ejemplo, comida hecha en el hogar o comprarla en el mercado. De esta forma pueden asignar tiempo a trabajar en su casa para consumir o en el mercado para comprarla allí)

Los bienes de consumo se pueden comprar en el mercado, en cantidad $C_M$, o producirlos en el hogar, en cantidad $C_D$, con $C = C_D + C_M$ (son sustitutos perfectos). Denotaremos: 

- $L_0$ la dotación total de tiempo disponible 
- $h_D$ el tiempo de trabajo en el hogar, $h_M$ tiempo de trabajo en el mercado 
- $C_D = f(h_D) la función de producción, con $f'\geq 0$ y $f'' \leq 0$ 
- $D = L_0 - h_D - h_M$ es la cantidad de ocio 

Entonces el problema del individuo es: 

$\displaystyle \max_{C_M, C_D, D, h_D, h_M} U(C,D)$ 

Sujeto a una serie de restricciones: 

1. $C_M \leq wh_M + R $
2. $C_D = f(h_D) $
3. $C = C_M + C_D $
4. $D = L_0 - h_D - h_M$ 

Entonces el problema se puede resumir en: 

$\displaystyle \max_{ h_D, h_M} U(wh_M + R + f(h_D),L_0 - h_M - h_D)$ 

Las soluciones óptimas de consumo y de ocio son: 

$ \dfrac{U_L(C*, D*)}{U_C(C*, D*)} = w = f'(h_D^*)$

$C* + wD* = \tilde{R_0^*}$

con el ingreso potencial $\tilde{R_0^*} = R + wL_0 + f(h_D^*) - wh_D^*$

Por consiguiente, el agente tiene un interés en dedicar su tiempo de trabajo a actividades domésticas en la medida en que la productividad marginal de una hora de este tipo de trabajo sea superior al salario horario.

Es interesante ver como la función de producción "total" parte, de izquierda a derecha, con solo producción en el hogar porque, gracias a la concavidad de la función de producción doméstica, la pendiente (transformación horas-consumo) es muy grande. Esto es así hasta $h_D$, donde las pendientes de las dos funciones de producción se igualan y el individuo está indiferente entre el hogar y el mercado, de ahí en mas la función de producción "total" es igual a la linea negra.
